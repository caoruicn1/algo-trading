## Display Pandas dataframe in a jQuery DataTable

Read the Titanic data set into a dataframe

```python
import pandas as pd
url = "https://raw.github.com/mattdelhey/kaggle-titanic/master/Data/train.csv"
titanic = pd.read_csv(url)
```

Create the DataTable function to insert a jquery DataTable

```python
import uuid

def DataTable(df):
    """ Prints a pandas.DataFrame as JQuery DataTables """
    from IPython.display import HTML
    # Generate random container name
    id_container = uuid.uuid1()
    output = """
        <div id="datatable-container-%s">
            <link rel="stylesheet" type="text/css" href="http://ajax.aspnetcdn.com/ajax/jquery.dataTables/1.9.0/css/jquery.dataTables.css">
            <script type="text/javascript" src="http://ajax.aspnetcdn.com/ajax/jquery.dataTables/1.9.0/jquery.dataTables.min.js"></script>

            <script type="text/javascript">
                $('#datatable-container-%s table.datatable').DataTable();
            </script>
            <!-- Insert table below -->
            %s
        </div>
    """ % (id_container, id_container, df.to_html(index=False, classes="datatable dataframe"))
    return HTML(output)
```

Display the dataframe in a jQuery DataTable

```python
DataTable(titanic[:600])
```
